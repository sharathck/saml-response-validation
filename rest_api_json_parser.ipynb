```python
import requests, json

try:
    response = requests.get('https://json-files-rest-api.s3.us-east-2.amazonaws.com/empcontactnested-with-root.json', timeout=5)
    print(response.text) #source file as it is printed
    json_response = json.loads(response.text)
    print(json.dumps(json_response,  indent=1)) #formatted source file
    print(json.dumps(json_response, sort_keys=True, indent=2)) #formatted source file with sorted keys - empid comes after contact info
    response.raise_for_status()
    # Code here will only run if the request is successful
except requests.exceptions.HTTPError as errh:
    print(errh)
except requests.exceptions.ConnectionError as errc:
    print(errc)
except requests.exceptions.Timeout as errt:
    print(errt)
except requests.exceptions.RequestException as err:
    print(err)
```
